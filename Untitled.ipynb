```python
import sympy
from sympy import cos, sin, sinh
from einsteinpy.symbolic import *

sympy.init_printing()
```

```python
coords = sympy.symbols("t r theta phi")
G, M, c, a = sympy.symbols("G M c a")
# using metric values of Kerr space-time

list2d = [[0 for i in range(4)] for i in range(4)]
list2d[0][0] = (2*M*coords[1] - coords[1]**2 - a**2 * sympy.cos(coords[2])**2)/ \
                (coords[1]**2 + a**2 * sympy.cos(coords[2])**2)
list2d[1][1] = (coords[1]**2 + a**2 * sympy.cos(coords[2])**2)/(a**2 - 2*M*coords[1] + coords[1]**2 )
list2d[2][2] = coords[1]**2 + a**2 * sympy.cos(coords[2])**2
list2d[3][3] = ((coords[1]**4 + a**2 * coords[1]**2 + 2*M*coords[1]*a**2 * sympy.sin(coords[2])**2 \
                 + (a**4 + a**2 * coords[1]**2)*sympy.cos(coords[2])**2 )/ \
                (coords[1]**2 + a**2 * sympy.cos(coords[2])**2))*sympy.sin(coords[2])**2
list2d[0][3] = -2*M*a*coords[1]*sympy.sin(coords[2])**2 /(coords[1]**2 + a**2 * sympy.cos(coords[2])**2)
list2d[3][0] = -2*M*a*coords[1]*sympy.sin(coords[2])**2 /(coords[1]**2 + a**2 * sympy.cos(coords[2])**2)
kerr = MetricTensor(list2d, coords)
kerr.tensor()
```

```python
%%time
kerr_chris = ChristoffelSymbols.from_metric(kerr)
kerr_chris.tensor()
```

```python
%%time
# Calculating Riemann Tensor from Christoffel Symbols
rm1 = RiemannCurvatureTensor.from_christoffels(kerr_chris)
rm1.tensor()
```
